```python
# ==========================================================
# Schema Validation Notebook
# ==========================================================

from pei_pipeline.main import main


# Parameters supplied by the Databricks Workflow.
dbutils.widgets.text(
    "pipeline_run_id",
    ""
)

dbutils.widgets.text(
    "attempt_id",
    ""
)

dbutils.widgets.text(
    "attempt_number",
    "1"
)


pipeline_run_id = (
    dbutils.widgets
    .get("pipeline_run_id")
    .strip()
)

attempt_id = (
    dbutils.widgets
    .get("attempt_id")
    .strip()
)

attempt_number_value = (
    dbutils.widgets
    .get("attempt_number")
    .strip()
)


if not pipeline_run_id:

    raise ValueError(
        "pipeline_run_id was not supplied. "
        "Schema Validation must use the pipeline_run_id "
        "of the Bronze ingestion batch."
    )


if not attempt_id:

    raise ValueError(
        "attempt_id was not supplied by the workflow."
    )


try:

    attempt_number = int(
        attempt_number_value
    )

except (TypeError, ValueError):

    raise ValueError(
        "attempt_number must be a positive integer. "
        f"Received: {attempt_number_value!r}"
    )


if attempt_number < 1:

    raise ValueError(
        "attempt_number must be greater than or equal to 1."
    )


result = main(
    spark=spark,
    dbutils=dbutils,
    stage="schema_validation",
    pipeline_run_id=pipeline_run_id,
    attempt_id=attempt_id,
    attempt_number=attempt_number
)

display(result)
```
