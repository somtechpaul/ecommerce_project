```python
# Databricks notebook source

# ==========================================================
# Enrichment Notebook
# ==========================================================

from pei_pipeline.main import main


# ==========================================================
# Workflow Parameters
# ==========================================================

dbutils.widgets.text(
    "pipeline_run_id",
    ""
)

dbutils.widgets.text(
    "attempt_id",
    ""
)

dbutils.widgets.text(
    "attempt_number",
    "1"
)


pipeline_run_id = (
    dbutils.widgets
    .get("pipeline_run_id")
    .strip()
)

attempt_id = (
    dbutils.widgets
    .get("attempt_id")
    .strip()
)

attempt_number_value = (
    dbutils.widgets
    .get("attempt_number")
    .strip()
)


# ==========================================================
# Validate Workflow Parameters
# ==========================================================

if not pipeline_run_id:

    raise ValueError(
        "pipeline_run_id was not supplied by the "
        "Databricks Workflow. Enrichment must use "
        "the same pipeline_run_id as Ingestion, "
        "Schema Validation and Data Quality."
    )


if not attempt_id:

    raise ValueError(
        "attempt_id was not supplied by the "
        "Databricks Workflow."
    )


try:

    attempt_number = int(
        attempt_number_value
    )

except (TypeError, ValueError) as error:

    raise ValueError(
        "attempt_number must be an integer. "
        f"Received: {attempt_number_value!r}"
    ) from error


if attempt_number < 1:

    raise ValueError(
        "attempt_number must be greater than or equal to 1."
    )


# ==========================================================
# Run Enrichment Pipeline
# ==========================================================

result = main(
    spark=spark,
    dbutils=dbutils,
    stage="enrichment",
    pipeline_run_id=pipeline_run_id,
    attempt_id=attempt_id,
    attempt_number=attempt_number
)

display(result)
```
